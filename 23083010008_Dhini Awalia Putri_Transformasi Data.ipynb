### Mengecek Data Outlier

```python
# Mendapatkan semua kolom numerik
numeric_columns = data.select_dtypes(include=['float64']).columns

# Membuat boxplot untuk setiap kolom numerik
plt.figure(figsize=(10, 15))
for i, column in enumerate(numeric_columns):
    plt.subplot(len(numeric_columns), 1, i + 1)
    sns.boxplot(y=data[column])
    plt.title(f'Boxplot untuk {column}')
    plt.xlabel(column)

plt.tight_layout()
plt.show()
```

### Mengatasi Data Outlier

```python
import pandas as pd
import scipy.stats as stats


columns = ['Pengeluaran per Kapita Disesuaikan (Ribu Rupiah/Orang/Tahun)', 'PDRB atas Dasar Harga Konstan menurut Pengeluaran (Rupiah)', 'Klasifikasi Kemiskinan']

for column in columns:
    data[column] = stats.mstats.winsorize(data[column], limits=[0.05, 0.05])

print(data[['Pengeluaran per Kapita Disesuaikan (Ribu Rupiah/Orang/Tahun)', 'PDRB atas Dasar Harga Konstan menurut Pengeluaran (Rupiah)', 'Klasifikasi Kemiskinan']].describe())

data.to_csv('data_winsorized.csv', index=False)
```
