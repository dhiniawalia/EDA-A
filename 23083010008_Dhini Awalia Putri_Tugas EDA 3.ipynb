### Deskripsi Kasus

Data yang digunakan merupakan data tingkat kemiskinan di Indonesia. Data ini mencakup beberapa indikator penting terkait kesejahteraan masyarakat, pembangunan manusia, kemiskinan, dan akses terhadap layanan dasar. Berikut adalah deskripsi dari setiap kolom yang ada di data tersebut:

1. Provinsi: Nama provinsi di mana kabupaten/kota tersebut berada. Dalam data ini, ada dua provinsi, yaitu Aceh dan Sumatera Utara.

2. Kab/Kota: Nama kabupaten atau kota yang merupakan unit administratif dalam provinsi.

3. Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen): Persentase penduduk yang hidup di bawah garis kemiskinan di masing-masing kabupaten/kota.

4. Rata-rata Lama Sekolah Penduduk 15+ (Tahun): Jumlah tahun rata-rata sekolah yang telah ditempuh oleh penduduk yang berusia 15 tahun ke atas di masing-masing wilayah.

5. Pengeluaran per Kapita Disesuaikan (Ribu Rupiah/Orang/Tahun): Rata-rata pengeluaran per kapita (disesuaikan) di masing-masing kabupaten/kota dalam satuan ribu rupiah per orang per tahun.

6. Indeks Pembangunan Manusia (IPM): Indeks yang mengukur pencapaian rata-rata dalam tiga dimensi dasar pembangunan manusia, yaitu kesehatan (diukur dari umur harapan hidup), pendidikan (rata-rata lama sekolah dan harapan lama sekolah), dan standar hidup layak (diukur dari PDB per kapita).

7. Umur Harapan Hidup (Tahun): Rata-rata tahun seseorang diperkirakan akan hidup sejak lahir di setiap kabupaten/kota.

8. Persentase Rumah Tangga yang Memiliki Akses terhadap Sanitasi Layak: Persentase rumah tangga yang memiliki akses terhadap sanitasi layak di masing-masing kabupaten/kota.

9. Persentase Rumah Tangga yang Memiliki Akses terhadap Air Minum Layak: Persentase rumah tangga yang memiliki akses terhadap air minum layak di masing-masing kabupaten/kota.

10. Tingkat Pengangguran Terbuka: Persentase angkatan kerja yang menganggur di setiap kabupaten/kota.

11. Tingkat Partisipasi Angkatan Kerja: Persentase jumlah penduduk yang aktif dalam angkatan kerja di setiap kabupaten/kota.

12. PDRB atas Dasar Harga Konstan Menurut Pengeluaran (Rupiah): Produk Domestik Regional Bruto (PDRB) yang diukur atas dasar harga konstan (dengan tahun acuan tertentu) menurut pengeluaran, yang merepresentasikan output ekonomi di kabupaten/kota tersebut.

13. Klasifikasi Kemiskinan: Indikator biner yang menunjukkan apakah suatu wilayah diklasifikasikan sebagai "Miskin" (1) atau "Tidak Miskin" (0) berdasarkan berbagai indikator kemiskinan.

## Pemusatan Data

```python
# Cleaning the data: replacing commas with dots and converting columns to numeric types for analysis
data_cleaned = data.copy()

# Replace commas with dots and convert columns to numeric
data_cleaned['Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)'] = pd.to_numeric(
    data_cleaned['Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)'].str.replace(',', '.')
)

data_cleaned['Indeks Pembangunan Manusia'] = pd.to_numeric(
    data_cleaned['Indeks Pembangunan Manusia'].str.replace(',', '.')
)

data_cleaned['Tingkat Pengangguran Terbuka'] = pd.to_numeric(
    data_cleaned['Tingkat Pengangguran Terbuka'].str.replace(',', '.')
)

data_cleaned['Tingkat Partisipasi Angkatan Kerja'] = pd.to_numeric(
    data_cleaned['Tingkat Partisipasi Angkatan Kerja'].str.replace(',', '.')
)

# 'Klasifikasi Kemiskinan' is already in numeric format
# Now, calculate the descriptive statistics for the selected variables
stats = data_cleaned[
    ['Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)',
     'Indeks Pembangunan Manusia',
     'Tingkat Pengangguran Terbuka',
     'Tingkat Partisipasi Angkatan Kerja']
].agg(['mean', 'median', 'std'])

stats
```

Perhitungan di atas adalah analisis deskriptif dari faktor-faktor yang mempengaruhi kemiskinan di suatu wilayah. Metode yang digunakan meliputi pencarian rata-rata (mean), nilai tengah (median), dan simpangan baku (standar deviasi) dari variabel-variabel yang menjadi faktor pengaruh. Dengan menganalisis rata-rata, median, dan standar deviasi dari faktor-faktor pengaruh ini, kita dapat memahami distribusi dan variasi dari kemiskinan, pembangunan manusia, pengangguran, dan partisipasi angkatan kerja di Indonesia. 

## Penyebaran Data

```python
# Definisikan kolom yang akan dianalisis
columns_for_analysis = [
    'Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)', 
    'Indeks Pembangunan Manusia', 
    'Tingkat Pengangguran Terbuka', 
    'Tingkat Partisipasi Angkatan Kerja'
]

# Buat dictionary untuk menyimpan hasil penyebaran data (variance dan standard deviation)
dispersion_data = {}

# Hitung varians dan standar deviasi untuk setiap kolom
for column in columns_for_analysis:
    variance = data_cleaned[column].var()
    std_dev = data_cleaned[column].std()
    dispersion_data[column] = {
        'Variance': variance,
        'Standard Deviation': std_dev
    }

# Konversi hasil ke DataFrame agar lebih mudah dibaca
dispersion_df = pd.DataFrame(dispersion_data).T

# Cetak hasilnya
print(dispersion_df)
```

Dari hasil tersebut, dapat dilihat bahwa variabel Persentase Penduduk Miskin dan Indeks Pembangunan Manusia memiliki penyebaran yang lebih besar dibandingkan variabel lain, yang berarti bahwa perbedaan antar wilayah dalam aspek kemiskinan dan pembangunan manusia lebih mencolok. Sedangkan Tingkat Pengangguran Terbuka memiliki penyebaran yang relatif kecil, menunjukkan bahwa kasus pengangguran cenderung lebih seragam di berbagai wilayah. Jika dilihat dari penyebaran data, variabel-variabel lain seperti Tingkat Pengangguran Terbuka dan Tingkat Partisipasi Angkatan Kerja menunjukkan penyebaran yang lebih kecil dan lebih seragam dibandingkan dengan variabel kemiskinan dan indeks pembangunan manusia.

## Analisis Skewness

```python
# Import pustaka yang diperlukan
import pandas as pd
from scipy.stats import skew

# Pilih kolom yang ingin dihitung skewness-nya
selected_columns = ['Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)', 
                    'Indeks Pembangunan Manusia', 
                    'Tingkat Pengangguran Terbuka', 
                    'Tingkat Partisipasi Angkatan Kerja']

# Filter data untuk hanya kolom yang dipilih
data_selected = data_cleaned[selected_columns]

# Buat dictionary untuk menyimpan hasil skewness
skewness_data = {}

# Loop melalui setiap kolom yang dipilih dan hitung skewness
for column in data_selected.columns:
    skew_value = skew(data_selected[column].dropna())  # dropna() untuk mengabaikan nilai yang hilang
    skewness_data[column] = skew_value

# Mengonversi hasil ke DataFrame agar lebih mudah dibaca
skewness_df = pd.DataFrame.from_dict(skewness_data, orient='index', columns=['Skewness'])

# Cetak hasil skewness
print(skewness_df)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, mode

# Pilih kolom yang ingin diplot
selected_columns = ['Persentase Penduduk Miskin (P0) Menurut Kabupaten/Kota (Persen)', 
                    'Indeks Pembangunan Manusia', 
                    'Tingkat Pengangguran Terbuka', 
                    'Tingkat Partisipasi Angkatan Kerja']

# Set ukuran plot
plt.figure(figsize=(16, 10))

# Loop untuk setiap kolom yang ingin ditampilkan
for i, column in enumerate(selected_columns):
    plt.subplot(2, 2, i+1)
    
    # Plot distribusi menggunakan seaborn
    sns.histplot(data_cleaned[column], kde=True, color='skyblue', stat='density')
    
    # Hitung mean, median, dan modus
    mean_value = data_cleaned[column].mean()
    median_value = data_cleaned[column].median()
    mode_value = mode(data_cleaned[column])[0][0]
    
    # Plot garis vertikal untuk mean, median, dan modus
    plt.axvline(mean_value, color='red', linestyle='--', label=f'Mean: {mean_value:.2f}')
    plt.axvline(median_value, color='green', linestyle='-', label=f'Median: {median_value:.2f}')
    plt.axvline(mode_value, color='orange', linestyle='-.', label=f'Modus: {mode_value:.2f}')
    
    # Beri judul pada subplot
    plt.title(f'{column} (Skewness: {skew(data_cleaned[column]):.2f})')
    
    # Tampilkan legenda
    plt.legend()

# Tampilkan seluruh plot
plt.tight_layout()
plt.show()
```

Analisis distribusi data dari empat variabel memberikan wawasan penting mengenai kondisi sosial dan ekonomi. Persentase Penduduk Miskin (P0) memiliki skewness 1.50 (condong ke kanan), dengan rata-rata 12.27, median 10.46, dan modus 4.62, menunjukkan bahwa sebagian besar kabupaten/kota memiliki persentase kemiskinan rendah, tetapi ada beberapa dengan angka tinggi. Indeks Pembangunan Manusia (IPM) menunjukkan skewness -0.78 (condong ke kiri), dengan rata-rata 69.93, median 69.61, dan modus 68.16, menunjukkan mayoritas wilayah memiliki nilai IPM tinggi meskipun ada yang rendah. Tingkat Pengangguran Terbuka memiliki skewness 0.79, dengan rata-rata 5.06, median 4.56, dan modus 5.86, menandakan mayoritas wilayah memiliki tingkat pengangguran rendah. Sementara itu, Tingkat Partisipasi Angkatan Kerja menunjukkan skewness 1.18, dengan rata-rata 69.46, median 68.95, dan modus 65.07, mencerminkan Tingkat Partisipasi Angkatan Kerja yang mendekati rata-rata di sebagian besar wilayah, tetapi ada beberapa yang lebih tinggi. Dari semua histogram tersebut, kita bisa melihat bagaimana skewness mempengaruhi distribusi data. Pada kasus di mana distribusi condong ke kanan (positif), nilai mean > median > modus. Sebaliknya, ketika distribusi condong ke kiri (negatif), median cenderung lebih besar dari mean. Skewness ini membantu dalam memprediksi pusat distribusi data dan perilaku variabel yang sedang dianalisis.
